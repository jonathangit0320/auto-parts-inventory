```python
query = ("SELECT * FROM LOCATION_WMS")

df = pd.read_sql_query(query, conn)
df
```

Contar Productos por Categoría Puedes contar cuántos productos hay en cada categoría

```python
def count_products_by_category(df):
    category_counts = df['category'].value_counts().reset_index()
    category_counts.columns = ['Category', 'Product_Count']
    return category_counts

# Llamar a la función y almacenar el resultado
category_summary = count_products_by_category(df)

# Imprimir el resultado
print(category_summary)
```

Total de Unidades por Ubicación Calcular el total de unidades almacenadas en cada ubicación

```python
units_per_location = df.groupby('location')['units'].sum().reset_index()
units_per_location.columns = ['Location', 'Total_Units']
print(units_per_location)
```

Filtrar Productos con Cantidad Baja Identificar productos con menos de un umbral específico de unidades

```python
def filter_low_stock_products(df, threshold=20):
    return df[df['units'] < threshold]

# Llamar a la función y almacenar el resultado
low_stock_products = filter_low_stock_products(df)

# Imprimir el resultado
print("Productos con Bajo Stock:")
print(low_stock_products[['item', 'product', 'units']])
```

Visualización de Productos por Categoría Crear un gráfico de barras para mostrar el número de productos por categoría

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='category', data=df)
plt.title('Número de Productos por Categoría')
plt.xticks(rotation=45)
plt.xlabel('Categoría')
plt.ylabel('Número de Productos')
plt.show()
```

Cantidad Total por Producto Calcular la cantidad total de unidades para cada producto

```python
total_units_per_product = df.groupby('product')['units'].sum().reset_index()
total_units_per_product.columns = ['Product', 'Total_Units']
print(total_units_per_product)
```
